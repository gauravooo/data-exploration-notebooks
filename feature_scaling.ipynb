```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler

df = pd.read_csv("train.csv")

features = df[["SalePrice", "GarageCars"]].dropna()
std_scaler = StandardScaler()
min_max_scaler = MinMaxScaler()

standardized_data = std_scaler.fit_transform(features)

normalized_data = min_max_scaler.fit_transform(features)

df_std = pd.DataFrame(standardized_data, columns=["SalePrice_Std", "GarageCars_Std"])
df_norm = pd.DataFrame(normalized_data, columns=["SalePrice_Norm", "GarageCars_Norm"])

print("---Original Data---")
print(features.head())

print("\n---Standardized Data (Mean=0, StdDev=1)---")
print(df_std.head().round(3))

print("\n---Standardized Data (Strictly 0 to 1)---")
df_norm.head().round(3)
```
