```python
df = pd.read_csv("train.csv")
df.columns
```

```python
df["Total_Home_SF"] = df["1stFlrSF"] + df["2ndFlrSF"] + df["TotalBsmtSF"]

df["House_age_at_Sale"] = df["YrSold"] - df["YearBuilt"]

df["Has_Pool"] = df["PoolArea"].apply(lambda x :1 if x>0 else 0)

print(df[["SalePrice","Total_Home_SF","House_age_at_Sale","Has_Pool"]].head())
```
